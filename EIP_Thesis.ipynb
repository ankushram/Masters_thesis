```python
import os
import cv2
import matplotlib.pyplot as plt


glioma_dir = '/content/drive/MyDrive/archive/preprocessed_testing/glioma'


glioma_images = os.listdir(glioma_dir)


glioma_images.sort()


first_image_path = os.path.join(glioma_dir, glioma_images[0])
image = cv2.imread(first_image_path, cv2.IMREAD_GRAYSCALE)


center_coordinates = (image.shape[1] // 2, image.shape[0] // 2)
radius = 50
color = (255, 0, 0)
thickness = 3


image_bgr = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)


circled_image = cv2.circle(image_bgr, center_coordinates, radius, color, thickness)


plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(circled_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Glioma MRI Image with Tumor Circled')
plt.show()


```
